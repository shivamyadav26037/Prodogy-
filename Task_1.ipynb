```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('/content/house_data.csv')  # Update this path as needed

# Group by number of bedrooms and compute average price
avg_price_by_bedrooms = df.groupby('bedrooms')['price'].mean()

# Plot bar chart
plt.figure(figsize=(8,5))
avg_price_by_bedrooms.plot(kind='bar', color='orange')
plt.title('Average House Price by Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Average Price')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
